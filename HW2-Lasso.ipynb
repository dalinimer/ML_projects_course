```python
import numpy as np
import matplotlib. pyplot as plt 
import pandas as pd
def a(X,k):
    return 2*(np.linalg.norm(X[:,k])**2)
 
def c(X,w,y,b,k):
    out=0
    for i in range(y.shape[0]):
        temp=0
        for j in range(w.shape[0]):
            if j!=k:
                temp+=w[j]*X[i,j]
        out+=(X[i,k]*(y[i]-(b+temp)))
    return 2*out
def w(c,a,Lambda):
    if c<-Lambda:
        return (c+Lambda)/a
    elif c>=-Lambda and c<=Lambda:
        return 0
    else:
        return (c-Lambda)/a
    
def lasso(X,y,Lambda,diff,w_init):
    n=X.shape[0]
    d=X.shape[1]
    new_w=w_init
    old_w=w_init
    delta=float("inf")
    while delta>diff:
        for i in range(new_w.shape[0]):
            old_w[i]=new_w[i]
        b=0
        for i in range(n):
            b+=y[i]-(new_w).dot(X[i,:])
        b=(1/n)*b
        for k in range(d):
            new_w[k]=w(c(X,new_w,y,b,k),a(X,k),Lambda)
        delta=np.linalg.norm(new_w-old_w)
       
    return new_w

def Lambda_max(X,Y):
    Y_m=np.mean(Y);
    Y=Y-Y_m
    maximum=0
    for k in range(X.shape[1]):
        maximum=max(np.transpose(X[:,k]).dot(Y),maximum)
    return 2*maximum
def plot_reg_path(X,Y,w_true,k):
    L=Lambda_max(X,Y)
    Lambdas=[]
    features=[]
    f=0
    w=[]
    fdr=[]
    tpr=[]
    print(w_true)
    while f<k:
        Lambdas.append(L)
        w=lasso(X,Y,L,0.1)
        y_temp=X.dot(w)
        fd=0
        tp=0
        for i in range(w.shape[0]):
            if w[i]!=0 and w_true[i]==0:
                fd+=1
            if w[i]!=0 and w_true[i]!=0:
                tp+=1
        if f!=0:
            fdr.append(fd/f)
            
        else:
            fdr.append(0)
        tpr.append((tp)/k)   
        f=np.count_nonzero(w)
        features.append(f)
        L=L/1.5
        
    plt.figure(1)
    plt.plot(Lambdas,features,'r')
    plt.xscale('log')
    plt.figure(2)
    plt.plot(fdr,tpr,'b')
    plt.figure(3)
    plt.plot(Lambdas,tpr,'b')
    plt.xscale('log')
    plt.figure(4)
    plt.plot(Lambdas,fdr,'b')
    plt.xscale('log')
    plt.show()

def synthetic_main():  
    k=50
    d=100
    n=100
    w_true=np.zeros((d,1))
    for j in range(k):
        w_true[j]=j/k
    X=np.random.normal(0,1,(n,d))
    Y=X.dot(w_true)+np.random.normal(0,1,(n,1))
    plot_reg_path(X,Y,w_true,k)

def data_main():
    df_train = pd.read_table("crime-train.txt")
    df_test = pd.read_table("crime-test.txt")
    Y_train=df_train["ViolentCrimesPerPop"].values
    X_train=df_train.drop("ViolentCrimesPerPop",axis=1).values
    Y_test=df_test["ViolentCrimesPerPop"].values
    X_test=df_test.drop("ViolentCrimesPerPop",axis=1).values
    L=Lambda_max(X_train,Y_train)
    Lambdas=[]
    features=[]
    f=0
    d=X_train.shape[1]
    w=np.zeros(d)
    W_1=[]
    W_2=[]
    W_3=[]
    W_4=[]
    W_5=[]
    MSE_test=[]
    MSE_train=[]
    while L>0.01:
        Lambdas.append(L)
        w=lasso(X_train,Y_train,L,0.1,w)
        features.append(np.count_nonzero(w))
        W_1.append(w[1])
        W_2.append(w[3])
        W_3.append(w[5])
        W_4.append(w[7])
        W_5.append(w[12])
        MSE_train.append(np.linalg.norm(X_train.dot(w)-Y_train)/(np.linalg.norm(Y_train)))
        MSE_test.append(np.linalg.norm(X_test.dot(w)-Y_test)/(np.linalg.norm(Y_test)))
        L=L/2
    w_30=lasso(X_train,Y_train,30,0.1,w)
    print("smallest feature",w_30.argmin())
    print("largest feature", w_30.argmax())
    plt.figure(1)
    plt.plot(Lambdas,features,'r')
    plt.xscale('log')
    plt.figure(2)
    plt.plot(Lambdas,W_1,'b')
    plt.xscale('log')
    plt.plot(Lambdas,W_2,'c')
    plt.xscale('log')
    plt.plot(Lambdas,W_3,'y')
    plt.xscale('log')
    plt.plot(Lambdas,W_4,'g')
    plt.xscale('log')
    plt.plot(Lambdas,W_5,'m')
    plt.xscale('log')
    plt.figure(3)
    plt.plot(Lambdas,MSE_train,'r')
    plt.xscale('log')
    plt.plot(Lambdas,MSE_test,'b')
    plt.xscale('log')

    

    
    
    
def main():
    df_train = pd.read_table("crime-train.txt")
    indexes=df_train.drop("ViolentCrimesPerPop",axis=1).columns.values
    print(indexes[38])
    print(indexes[45])
    
    
    
    
    
if __name__ == "__main__":
    main()
    
    
```
