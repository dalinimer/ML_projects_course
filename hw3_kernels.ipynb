```python
import numpy as np
import math

def create_inputs(n):
    return np.random.uniform(0,1,(n,1))
def f(X):
    return 4*np.sin(math.pi*X)*np.cos(6*math.pi*np.square(X))
def create_outputs(X,func):
    n=X.shape[0]
    return func(X)+np.random.normal(0,1,(n,1))
def kernel_matrix(kernel_func,X):
    n=X.shape[0]
    K=np.zeros((n,n))
    for i in range(n):
        for j in range(n):
            K[i,j]=kernel_func(X[i,0],X[j,0])
    return K
def alpha(X,Y,K,Lambda):
    return np.linalg.inv(K+Lambda*np.identity(X.shape[0])).dot(Y)
def predictor(alpha,kernel_func,X,x):
    temp=0
    for i in range(X.shape[0]):
        temp=temp+alpha[i,:]*kernel_func(X[i,:],x)
    return temp

def LOO_cross_V_poly(X,Y,n):
    MSE=np.zeros((n**2,1))
    d=np.zeros((n**2,1))
    
    Lambda=np.zeros((n**2,1))
    count=-1
    for i in range(n):
        for j in range(n):
            count=count+1
            k=lambda w,z: (1+w*z)**(j+1)
            Lambda[count,0]=(1)*(2**i)
            d[count,0]=j+1
            for l in range(n):
                X_train=np.delete(X,l,axis=0)
                Y_train=np.delete(Y,l,axis=0)
                x=X[l,0]
                K=kernel_matrix(k,X_train)
                a=alpha(X_train,Y_train,K,i+1)
                f=predictor(a,k,X_train,x)
                MSE[count,0]=MSE[count,0]+((f-Y[l,0]))**2
            MSE[count,0]=MSE[count,0]/n
    print(MSE)
    min_arg=np.argmin(MSE)
    print(min_arg)
    return [Lambda[min_arg,],d[min_arg,]]
        
        
def main():
    n=30
    X=create_inputs(n)
    Y=create_outputs(X,f)
    print(LOO_cross_V_poly(X,Y,n))
    
    
if __name__ == "__main__":
    main()
    
```
