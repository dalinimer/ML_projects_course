```python
import numpy as np
import matplotlib.pyplot as plt
def main():
    train_n=5
    test_n=1000
    d=5
    final=7
    repet=1
    for i in range(repet):
        X_train=np.random.normal(0,1,size=(train_n,d))
        a_true=np.random.normal(0,1,size=(d,1))
        y_train=X_train.dot(a_true)+np.random.normal(0,0.5,size=(train_n,1))
        X_test=np.random.normal(0,1,size=(test_n,d))
        y_test=X_test.dot(a_true)+np.random.normal(0,0.5,size=(test_n,1))
        W_min=np.zeros((final,d))
        lambda_v=0.0001
        index=0
        error_Train=np.zeros(final)
        error_Test=np.zeros(final)
        while index<final:
            W=np.transpose(X_train).dot(X_train)
            W=W+np.identity(d)*lambda_v
            print(np.identity(d)*lambda_v)
            W=np.linalg.inv(W)
            print(X_train)
            print(np.transpose(X_train).dot(X_train))
            print(np.transpose(X_train).dot(X_train)+np.identity(d)*lambda_v)            
            W_temp=W.dot(np.transpose(X_train).dot(y_train))
            W_min[index:index+1,:]=np.transpose(W_temp)
            error_Train[index]=+(np.linalg.norm(np.dot(X_train,np.transpose(W_min[index,:]))-y_train)/np.linalg.norm(y_train))
            error_Test[index]+=(np.linalg.norm(np.dot(X_test,np.transpose(W_min[index,:]))-y_test)/np.linalg.norm(y_test))
            index=index+1
            lambda_v=lambda_v*10
    error_Train=error_Train/repet
    error_Test=error_Test/repet
    plt.plot(error_Train,"b",label="error training")
    plt.plot(error_Test,"r",label="error testing")
    plt.show
    
        
if __name__ == "__main__":
    main()
    print("hi hi hi")



```
