```python
from mnist import MNIST
import numpy as np
import math as math
import matplotlib.pyplot as plt
import random as random

def L(w,b,X,Y,Lambda):
    y=-np.log(mu(w,b,X,Y))
    return np.average(y)+Lambda*(np.linalg.norm(w)**2)
    
def mu(w,b,X,y):
    Y=-np.multiply(y,b+X.dot(w))
    return np.divide(1,1+np.exp(Y))

def new_w(old_w,Y,X,b,Lambda,h):
    d=old_w.shape[0]
    w=np.copy(old_w)
    for j in range(d):
        x=np.transpose(X).dot((1-mu(w,b,X,Y))*Y)
        w[j,0]=w[j,0]+h*(np.average(x)-2*Lambda*w[j,0])
    return w
def new_b(b,X,Y,h,w):
    y=Y*(1-mu(w,b,X,Y))
    return b+h*(np.average(y))

def logistic_descent(w,b,X,Y,h,delta,Lambda,iteration_num,loss,batch):
    threshold=1000
    
    oldb=b
    newb=b
    count=0
    indexes=random.sample(range(0,X.shape[0]), batch)
    X_stoch=X[indexes,:]
    Y_stoch=Y[indexes]
    oldw=w
    neww=w
    
    while threshold>delta:
        newb=new_b(oldb,X_stoch,Y_stoch,h,neww)
        neww=new_w(oldw,Y_stoch,X_stoch,b,Lambda,h)
        count+=1
        threshold=(np.linalg.norm(neww-oldw))**2 + (abs(newb-oldb)**2)
        
        
    
    loss.append(L(neww,newb,X_stoch,Y_stoch,Lambda))
    iteration_num.append(count)
    print(loss)
    print(iteration_num)

def load_dataset():
    mndata = MNIST("./data/")
    X_train, labels_train = map(np.array, mndata.load_training())
    X_test, labels_test = map(np.array, mndata.load_testing())
    X_train = X_train/255.0
    X_test = X_test/255.0
   
    labels_train_binary=[]
    X_train_binary=[[]]
    labels_test_binary=[]
    X_test_binary=[[]]
  
    for i in range(labels_train.shape[0]):
        if labels_train[i]==2:
            labels_train_binary.append(-1)
        if labels_train[i]==7:
            labels_train_binary.append(1)
        
            
    for i in range(labels_test.shape[0]):
        if labels_test[i]==2:
            labels_test_binary.append(-1)
        if labels_test[i]==7:
            labels_test_binary.append(1)
        
    d=X_train.shape[1]
    X_train_st=np.zeros((len(labels_train_binary),d))
    X_test_st=np.zeros((len(labels_test_binary),d))
    for i in range(labels_train.shape[0]):
        X_train_st[i:i+1,:]=X_train[i:i+1,:]
    for i in range(labels_test.shape[0]):
        X_test_st[i:i+1,:]=X_test[i:i+1,:]
    labels_train_st=np.asarray(labels_train_binary)
    labels_test_st=np.asarray(labels_test_binary)
    iteration_num=[]
    loss=[]
    w=np.zeros((d,1))
    b=0
    h=10
    for i in range(5):
        logistic_descent(w,b,X_train_st,labels_train_st,h,0.5,0.1,iteration_num,loss,1)
        h=h/2
    plt.figure(1)
    plt.plot(iteration_num,loss,'r')
    
    
def main():
    load_dataset()
if __name__ == "__main__":
    print("hi")
    main()         
```
